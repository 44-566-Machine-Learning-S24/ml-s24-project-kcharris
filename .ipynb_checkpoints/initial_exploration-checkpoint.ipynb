Budget is shown as an int64 type but is not listed in the describe, so I will manually change the data, as well as create a work copy of the data to modify.

```python
work_copy = data_frame.copy()
work_copy['budget'] = pd.to_numeric(work_copy['budget'], errors='coerce')
work_copy.describe()
```

```python
import matplotlib.pyplot as plt

work_copy.hist(bins=50)
plt.tight_layout()
plt.show()
```

```python
from pandas.plotting import scatter_matrix
compare = ['budget', 'revenue', 'runtime', 'vote_average']
scatter_matrix(work_copy[compare])
```

```python
c = work_copy[compare].corr()
print(c)
```
