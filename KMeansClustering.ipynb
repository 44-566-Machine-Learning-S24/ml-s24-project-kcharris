```python
import pandas as pd
from ast import literal_eval
from sklearn.model_selection import train_test_split

# read from file
data_frame = pd.read_csv("../ProjectResources/movies/movies_metadata.csv", low_memory=False)

#create working copy
work_copy = data_frame.copy()
work_copy['budget'] = pd.to_numeric(work_copy['budget'], errors='coerce')

# clean the data
work_copy.dropna(axis='index', how='any', subset=['budget','vote_average', 'revenue', 'genres'], inplace=True)
work_copy = work_copy[['budget','genres', 'vote_average', 'revenue']]
# drop data with bad values
work_copy = work_copy.drop(work_copy[work_copy.revenue == 0].index)

# object under genres can have multiple values, expand the dataset by creating multiple rows for the
# same movie but with a single genre. These single genre items will be in {id: #, name: "genre_name"} format
work_copy['genres'] = work_copy['genres'].apply(literal_eval)
work_copy = work_copy.explode("genres")
work_copy.dropna(axis='index', how='any', subset=['genres'], inplace=True)
work_copy['genres'] = [d.get('name') for d in work_copy['genres']]

# Encode the genres into integers so they can be used by the KMeans classifier.
from sklearn.preprocessing import LabelEncoder
work_copy['genre_int'] = LabelEncoder().fit_transform(work_copy['genres'])

# create training and test sets
train_set, test_set = train_test_split(work_copy, test_size=0.2, random_state=777)
```

```python
# verify it works
# label encoder starts at 0. Cluster should concider number of genres
num_genres = max(work_copy["genre_int"])+1
print(f"The number of genres is {num_genres}")
work_copy.head(10)
```

Try clustering with revenue and the 20 genres

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, f1_score
X = train_set[['revenue', 'genre_int']]

kmeans = KMeans(n_clusters=20, n_init=1000)
kmeans.fit(X)
```

```python
import math
RMS_average = math.sqrt(kmeans.inertia_ / len(X))
print(f"The genre cluster distortion is {kmeans.inertia_} RMS {RMS_average}")
print(f"The genre centroids are: \n{kmeans.cluster_centers_}")
```

```python
import matplotlib.pyplot as plt

cluster_x = [x for [x, y] in kmeans.cluster_centers_]
cluster_y = [y for [x, y] in kmeans.cluster_centers_]

plt.scatter(train_set["revenue"], train_set['genre_int'], c = kmeans.labels_)
plt.scatter(cluster_x, cluster_y, marker="+", color = "red")
plt.xlabel("revenue")
plt.ylabel("genre_int")
# scaling results in a graph showing nothing
# plt.axis("scaled")
plt.savefig("fig/kmeans")
plt.show()
```
